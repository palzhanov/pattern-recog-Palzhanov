```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 18})
```

```python
n = 100 #matrix size
a = np.random.randn(n, n)
a.shape
```

```python
U, S, VT = np.linalg.svd(a, full_matrices=False)
plt.plot(S, '.', label = 'Singular Values')
plt.legend()
plt.show()
```

```python
SV = np.zeros([100,n])
for i in range(100):
    a = np.random.randn(n, n)
    U, S, VT = np.linalg.svd(a, full_matrices=False)
    SV[i,:] = S
```

```python
plt.figure(figsize =(10, 7))
plt.boxplot(SV[:,:15]) #first 15 singular values for better view
plt.show()
```

```python
#mean and median of singular values
means = np.zeros(n)
medians = np.zeros(n)
xlab = np.linspace(1,n,n)
for i in range(n):
    means[i] = np.mean(SV[:,i])
for i in range(n):
    medians[i] = np.median(SV[:,i])
fig = plt.figure(figsize =(8, 8))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(xlab, means, 'x')
ax.plot(xlab, medians, '.')
plt.show()
```
